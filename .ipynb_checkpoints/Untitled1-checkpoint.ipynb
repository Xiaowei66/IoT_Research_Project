```python
# -*- coding: utf-8 -*- 
# @Time : 2019/6/6 8:13 
# @Author : WangPeng
from sympy import *
from math import radians, cos, sin, asin, sqrt


class Get_new_gps():
    def __init__(self):
        # 地球半径
        self.R = 6371 * 1000
        pass

    """计算两点间距离"""

    def geodistance(self, lng1, lat1, lng2, lat2):
        """

        :param lng1: 120.12802999999997
        :param lat1: 30.28708,115
        :param lng2: 115.86572000000001
        :param lat2: 28.7427
        :return: 两点间距离
        """
        lng1, lat1, lng2, lat2 = map(radians, [float(lng1), float(lat1), float(lng2), float(lat2)])  # 经纬度转换成弧度
        dlon = lng2 - lng1
        dlat = lat2 - lat1
        a = sin(dlat / 2) ** 2 + cos(lat1) * cos(lat2) * sin(dlon / 2) ** 2
        distance = 2 * asin(sqrt(a)) * self.R  # 地球平均半径，6371km
        distance = round(distance, 3)
        return distance

    """计算点经纬度北500米的点的经纬度"""

    def get_new_lat(self, lng1, lat1, dist=500):
        """

        :param lng1: 120.12803
        :param lat1: 20.28708
        :param dist:
        :return: (120.12803, 20.291576608029594)
        """
        lat2 = 180 * dist / (self.R * pi) + lat1
        return (lng1, lat2)

    """计算点经纬度正东500米点的经纬度"""

    def get_new_lng(self, lng1, lat1, dist=500):
        """

        :param lng1: 116.55272514141352
        :param lat1: 30.28708
        :param dist:
        :return: (116.55272514141352,30.28708)
        """
        lng2 = 180 * dist / (self.R * pi * cos(radians(lat1))) + lng1
        return (lng2, lat1)

    """计算点东北方向与正东某夹角某距离的经纬度"""

    def get_new_lng_angle(self, lng1, lat1, dist=500, angle=30):
        """

        :param lng1:116.55272514141352
        :param lat1:30.28708
        :param dist:指定距离
        :param angle:指定角度
        :return:（0.0091871843081617/pi + 116.498079 0.0122339171779312/pi + 39.752304）
        """
        lat2 = 180 * dist*sin(radians(angle)) / (self.R * pi) + lat1
        lng2 = 180 * dist*cos(radians(angle)) / (self.R * pi * cos(radians(lat1))) + lng1
        return (lng2, lat2)


functions = Get_new_gps()

if __name__ == '__main__':
    functions = Get_new_gps()
    lng1, lat1 = [151.181624,-33.794506]
    # 计算正北的点
    lng2, lat2 = functions.get_new_lat(lng1, lat1)
    # 计算正东的点
    lng3, lat3 = functions.get_new_lng(lng1, lat1)
    # 计算该点与正东逆时针夹角45度，距离500米点的经纬度值
    lng4, lat4 = functions.get_new_lng_angle(lng1, lat1, dist=69, angle=335)
    print(lng4, lat4)
    print("原始点的经纬度坐标", lng1, lat1)
    print("正北500米坐标点为%f,%f,距离计算为%f米" % (lng2, lat2, functions.geodistance(lng1, lat1, lng2, lat2)))
    print("正东500米坐标点为%f,%f,距离计算为%f米" % (lng3, lat3, functions.geodistance(lng1, lat1, lng3, lat3)))
    print("东北方夹角，距离500米坐标点为%f,%f,距离计算为%f米" % (lng4, lat4, functions.geodistance(lng1, lat1, lng3, lat3)))
    print(functions.geodistance(lng3, lat3, lng4, lat4))
```

```python

import geopy as g 
import geopy.distance as gp 
from geopy.units import radians
import math

#################################################################
from sympy import *
from math import radians, cos, sin, asin, sqrt

class Get_new_gps():
    def __init__(self):
        # 地球半径
        self.R = 6371 * 1000
        pass

    """计算两点间距离"""

    def geodistance(self, lng1, lat1, lng2, lat2):
        """

        :param lng1: 120.12802999999997
        :param lat1: 30.28708,115
        :param lng2: 115.86572000000001
        :param lat2: 28.7427
        :return: 两点间距离
        """
        lng1, lat1, lng2, lat2 = map(radians, [float(lng1), float(lat1), float(lng2), float(lat2)])  # 经纬度转换成弧度
        dlon = lng2 - lng1
        dlat = lat2 - lat1
        a = sin(dlat / 2) ** 2 + cos(lat1) * cos(lat2) * sin(dlon / 2) ** 2
        distance = 2 * asin(sqrt(a)) * self.R  # 地球平均半径，6371km
        distance = round(distance, 3)
        return distance

    """计算点经纬度北500米的点的经纬度"""

    def get_new_lat(self, lng1, lat1, dist=500):
        """

        :param lng1: 120.12803
        :param lat1: 20.28708
        :param dist:
        :return: (120.12803, 20.291576608029594)
        """
        lat2 = 180 * dist / (self.R * pi) + lat1
        return (lng1, lat2)

    """计算点经纬度正东500米点的经纬度"""

    def get_new_lng(self, lng1, lat1, dist=500):
        """

        :param lng1: 116.55272514141352
        :param lat1: 30.28708
        :param dist:
        :return: (116.55272514141352,30.28708)
        """
        lng2 = 180 * dist / (self.R * pi * cos(radians(lat1))) + lng1
        return (lng2, lat1)

    """计算点东北方向与正东某夹角某距离的经纬度"""

    def get_new_lng_angle(self, lng1, lat1, dist=500, angle=30):
        """

        :param lng1:116.55272514141352
        :param lat1:30.28708
        :param dist:指定距离
        :param angle:指定角度
        :return:（0.0091871843081617/pi + 116.498079 0.0122339171779312/pi + 39.752304）
        """
        lat2 = 180 * dist*sin(radians(angle)) / (self.R * pi) + lat1
        lng2 = 180 * dist*cos(radians(angle)) / (self.R * pi * cos(radians(lat1))) + lng1
        return (lng2, lat2)

################################################################
# This is the structure of basement
lable_meter = 3.5
number_of_rows = 17
number_of_columns = 14
# the conversed coordinates (compared with google map)
coordiantes_of_start_point = [151.181655, -33.794509]

# define a function that output the coordinates with the input of distance and angle
def get_coordinates(start_ponit, dist, angle):
    lng1,lat1 = start_ponit
    
    functions = Get_new_gps()
    
    lng4, lat4 = functions.get_new_lng_angle(lng1, lat1, dist, angle)
    print("Current coordinates:%f,%f"%(lat4, lng4))
    ans = [lat4, lng4]
    
    return ans

# define a function to get the disctance from current point to label -2
def get_distance(start_label,current_label):
    horizontal_len = lable_meter*(current_label//number_of_rows)
    vertical_len = lable_meter*((current_label-start_label)%number_of_rows)
    
    distance = math.sqrt(horizontal_len*horizontal_len + vertical_len*vertical_len)
    print("The distance to the reference point:",distance)
    return distance


def get_angle(start_label,current_label):
    horizontal_len = (current_label-start_label)//number_of_rows
    vertical_len = (current_label-start_label)%number_of_rows
    distance = math.sqrt(horizontal_len*horizontal_len + vertical_len*vertical_len)
    
    #print(horizontal_len,vertical_len,distance)
    if (horizontal_len == distance):
        angle = 21.5
    else:
        A=math.degrees(math.acos((horizontal_len*horizontal_len-vertical_len*vertical_len-distance*distance)/(-2*vertical_len*distance)))
        A = round(A)
        B = 90-A
        
        if (B <= 21.5):
            angle = 21.5-B
        else:
            angle = 360-(B-21.5)
    
    print("The angle of this point:",angle)
    return angle



def output_current_coordiantes(start_label,current_label):
    
    dist = get_distance(start_label,current_label)
    angle = get_angle(start_label,current_label)
    
    current_coordinates = get_coordinates(coordiantes_of_start_point, dist, angle)
    
    return current_coordinates

# (-2 means: the label that you used to refer, 232: means the label that you want to get coordinates)
# in this project we refer label -2 and it's coordinates is coordiantes_of_start_point = [-33.794508, 151.181634]
```
